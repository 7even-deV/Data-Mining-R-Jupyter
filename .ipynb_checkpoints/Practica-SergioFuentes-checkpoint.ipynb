Existen varios Barrios que todas sus entradas de Square.Meters son NA, vamos a eliminar del dataset todos los pisos que pertenecen a estos barrios.

```r
neighbourhood_NA <- df_madrid %>% mutate(Square.Meters.Na = ifelse(is.na(Square.Meters), 1, 0)) %>%
                    group_by(Neighbourhood) %>% summarise(nas=sum(Square.Meters.Na) / length(Square.Meters))
neighbourhood_NA <- neighbourhood_NA %>% filter(nas==1) %>% select(Neighbourhood)
neighbourhood_NA <- unique(neighbourhood_NA $Neighbourhood)

paste0("El n√∫mero de barrios con todas las entradas de Square.Meters NA es de ", length(unique(neighbourhood_NA)))
```

```r
df_madrid <- df_madrid[!df_madrid $Neighbourhood %in% neighbourhood_NA,]
head(df_madrid)

paste("De esta forma se mantienen", length(unique(df_madrid$Neighbourhood)), "barrios.")
```

El barrio parece ser un indicador importante para los metros cuadrados de un apartamento.

Vamos a agrupar los barrios por metros cuadrados. Podemos usar una matriz de similaridad de Tukey tal y como hicimos en el curso de estad√≠stica:

```r
tky <- TukeyHSD(aov( formula=Square.Meters ~ Neighbourhood, data=df_madrid ))
tky.result <- data.frame(tky $Neighbourhood)
cn <-sort(unique(df_madrid$Neighbourhood))
resm <- matrix(NA, length(cn),length(cn))
rownames(resm) <- cn
colnames(resm) <- cn
resm[lower.tri(resm) ] <- round(tky.result $p.adj, 4)
resm[upper.tri(resm) ] <- t(resm)[upper.tri(resm)]
diag(resm) <- 1
library(ggplot2)
library(reshape2)
dfResm <- melt(resm)
ggplot(dfResm, aes(x=Var1, y=Var2, fill=value))+
  geom_tile(colour = "black")+
  scale_fill_gradient(low = "white",high = "steelblue")+
  ylab("Class") + xlab("Class") + theme_bw() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1), legend.position="none")
```

Usando como variable de distancia: 1-resm
Dibuja un dendrograma de los diferentes barrios.


```r
dist <- as.dist(1 - resm)
hc <- hclust(dist, method="complete")
hcd <- as.dendrogram(hc)
plot(hcd)
```

```r
library(dendextend)

options(repr.plot.height=6, repr.plot.width=6, repr.plot.res=200)
hcd <- as.dendrogram(hc)
hcd <- set(hcd, "labels_cex", 0.4)
plot(color_branches(hcd, h=0.4), horiz=TRUE)
abline(v=0.4, col="red")
```

¬øQue punto de corte ser√≠a el aconsejable?, ¬øcuantos clusters aparecen?

```r
options(repr.plot.height=4, repr.plot.width=4, repr.plot.res=300)
cutree <- cutree(hc, h=0.4)

library(cluster)
ss <- silhouette(cutree, dist)
plot(ss, border=NA)
```

Usar PCA para encontrar el apartamento m√°s cercano a uno dado.

Este algoritmo nos ayudar√≠a a dado un apartamento que el algoritmo nos devolver√≠a los 5 apartamentos m√°s similares.


Crearemos una funci√≥n tal que le pasemos un apartamento con los siguientes datos:
* Accommodates	
* Bathrooms	
* Bedrooms	
* Beds	
* Price	
* Guests.Included	
* Extra.People	
* Review.Scores.Rating	
* Latitude	
* Longitude	
* Square.Meters

y nos devuelva los 5 m√°s similares de:

```r
df_madrid_pca <- df_madrid_original %>% dplyr :: select(
    c(Accommodates, Bathrooms, Bedrooms, Latitude, Longitude, Beds, Price, Review.Scores.Rating,
    Square.Meters, Neighb_id)) %>% na.omit()

head(df_madrid_pca)

pca_madrid <- prcomp(dplyr :: select(df_madrid_pca, -Neighb_id), center=TRUE, scale.=TRUE)
```

```r
plot(pca_madrid $sdev ^2 / sum(pca_madrid $sdev ^2), main="Autovalores")
str(pca_madrid)
```

```r
id_piso <- 17
piso <- df_madrid_pca[id_piso,]
pca_piso <- predict(pca_madrid, piso)
pca_older <- pca_madrid $x[,1:2]
pca_piso <- pca_piso[,1:2]

id_pca_result <- order(rowSums((pca_piso - pca_older) ^2))

df_madrid_pca[id_piso,]
df_madrid_pca[id_pca_result %in% 1:5,]
```
